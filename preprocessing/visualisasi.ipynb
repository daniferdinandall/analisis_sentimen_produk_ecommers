```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labeled_file = 'label/Aerostreet1_labeled.csv'
df = pd.read_csv(labeled_file)
```

```python
# Menghitung jumlah masing-masing label
label_counts = df['Label'].value_counts()
```

```python
# Mengatur gaya untuk seaborn
sns.set(style="whitegrid")
```

```python
# Membuat Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title('Sentiment Analysis Results - Bar Chart')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Membuat Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(label_counts.values, labels=label_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(label_counts)))
plt.title('Sentiment Analysis Results - Pie Chart')
plt.show()
```
