```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv('../Dataset/dataset_2.csv')
```

```python
# Menghitung jumlah masing-masing label
sentiment_counts = df['Label'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Jumlah Ulasan Berdasarkan Sentimen')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah Ulasan')
plt.show()
```

```python
# Membuat Word Cloud untuk setiap sentimen
for sentiment in df['Label'].unique():
    # Filter teks ulasan berdasarkan sentimen
    label_text = ' '.join(df[df['Label'] == sentiment]['Ulasan'])
    
    # Membuat Word Cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(label_text)
    
    # Menampilkan Word Cloud
    plt.figure(figsize=(15, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud untuk Ulasan: {sentiment}', fontsize=20)
    plt.show()
```

```python
puas_text = ' '.join(df[df['Label'] == 'Puas']['Ulasan'])
puas_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(puas_text)

plt.figure(figsize=(15, 8))
plt.imshow(puas_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud untuk Sentimen: Puas')
plt.show()
```
